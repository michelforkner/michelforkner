```python
# Extract articles
articles = ts["results"]

# Create list of dictionaries with useful fields
movies_data = []

for article in articles: # creating a dictionary to store information about one article
    movie = {
        "title": article["title"],
        "abstract": article["abstract"],
        "published_date": article["published_date"],
        "url": article["url"]
    }
    movies_data.append(movie) # appending each movie article to the movies_data list

# Convert to DataFrame (make data easier to work with, analyze, and export)
df = pd.DataFrame(movies_data)

print(df.head()) #first 5 rows

# Save to CSV so the data can be used outside Python
df.to_csv("nyt_movies_top_stories.csv", index=False) # Pandas normally adds an extra numbered column called the index, index=False prevents that extra column
```
